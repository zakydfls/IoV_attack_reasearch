```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('binary_all.csv')
df
```

```python
df = df.drop(columns=[col for col in df.columns if 'ID' in col or col in ['label', 'category']])
df
```

```python
label_encoder = LabelEncoder()
df['specific_class'] = label_encoder.fit_transform(df['specific_class'])
df['specific_class']
```

```python
X = df.drop(columns=['specific_class'])
y = df['specific_class']
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
mi_scores = mutual_info_classif(X, y)
mi_scores_series = pd.Series(mi_scores, index=X.columns)
```

```python
ranked_features = mi_scores_series.sort_values(ascending=False)
print("Feature Rank:")
print(ranked_features)
```
